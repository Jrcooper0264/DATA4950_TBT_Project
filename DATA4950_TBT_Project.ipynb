```python
import pandas as pd
```

```python
#Reading in data from IPEDS database

#will need to change working directory to where the data is stored
df = pd.read_csv('C:/Users/16154/Downloads/11-14-2022---860/data.csv')
```

```python
#Dropping all school that do not have a price estimate
df_r = df.dropna(subset = ['Total price for in-district students living on campus  2004-2005 (IC2004_AY)'])
```

```python
#pulling in the response data from the csv generated by the Google form
#https://docs.google.com/forms/d/e/1FAIpQLSdPZ8FlZOER39O4jxj7g6oVqq4wB4NRyBtQSVBKOv_EYYflPQ/viewform?usp=sf_link

#will need to change working directory to where the data is stored
df_response = pd.read_csv('C:/Users/16154/Downloads/Responses.csv')
```

```python
#Generating reports for each response batch

for i in range(0, len(df_response['What state(s) were you considering going to school?'])):
    
    Base_col = ['Institution Name', '12-month unduplicated headcount  total: Academic year 2004-05 (DFR2005_RV)',
               'State abbreviation (HD2004)', 'City location of institution (HD2004)', 'Street address or post office box (HD2004)',
               'ZIP code (HD2004)', 'Open admission policy (IC2004)', 'Secondary school GPA (IC2004)', 'Recommendations (IC2004)',
               'Admission test scores (IC2004)','Total price for in-district students living on campus  2004-2005 (IC2004_AY)',
               'Total price for out-of-state students living on campus 2004-2005 (IC2004_AY)',
               'Total price for in-state students living on campus 2004-2005 (IC2004_AY)']

    test_score = [ 'SAT I Verbal 25th percentile score (IC2004)',
                'SAT  I Verbal 75th percentile score (IC2004)', 'SAT I Math 25th percentile score (IC2004)', 
               'SAT I Math 75th percentile score (IC2004)', 'ACT Composite 25th percentile score (IC2004)',
                 'ACT Composite 75th percentile score (IC2004)', 'ACT English 25th percentile score (IC2004)',
                 'ACT English 75th percentile score (IC2004)', 'ACT Math 25th percentile score (IC2004)', 
                 'ACT Math 75th percentile score (IC2004)']

    Athletics = ['NCAA/NAIA member for football (IC2004)', 'NCAA/NAIA member for basketball (IC2004)',
                'NCAA/NAIA member for baseball (IC2004)', 'NCAA/NAIA member for cross country/track (IC2004)']

    final_col = Base_col

    if df_response['Did you take the ACT and/or SAT?'][i] == 'Yes':
        for item in test_score:
            final_col.append(item)

    if df_response['Are you planning on participating in one of the following collegiate sports?'][i] == 'Football':
        final_col.append(Athletics[0])
    elif df_response['Are you planning on participating in one of the following collegiate sports?'][i] == 'Basketball':
        final_col.append(Athletics[1])
    elif df_response['Are you planning on participating in one of the following collegiate sports?'][i] == 'Baseball':
        final_col.append(Athletics[2])
    elif df_response['Are you planning on participating in one of the following collegiate sports?'][i] == 'Cross Country/track':
        final_col.append(Athletics[3])

    df_recommend = df_r[final_col]
    
    state_list = df_response['What state(s) were you considering going to school?'][i]
    state_list = state_list.split(',')
    df_final = test = df_recommend.loc[df_recommend['State abbreviation (HD2004)'] == state_list[0]]
    for item in state_list[1:]:
        item = item.replace(" ", "")
        test = df_recommend.loc[df_recommend['State abbreviation (HD2004)'] == item]
        df_final = df_final.append(test, ignore_index=True)
    
    filename = "out%d.csv" % (i)
    df_final.to_csv(filename,index=False)
```
